```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
make_blobs?
X,y=make_blobs(n_samples=50, centers=3, n_features=2, random_state=2) #kmeans 그림 참조
print(X)
print(X.shape)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.title('Random data - no label')
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.title('Random data - with labels')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)  # X만 modeling and learning
```

```python
print(kmeans.labels_)
print(y)
print(kmeans.cluster_centers_)
```

```python
acc=np.sum(np.array(kmeans.labels_ == y))/y.size
print(acc)
```

```python
colors=['blue','black','green']
for color, i in zip(colors, [0,1,2]):
    plt.scatter(X[kmeans.labels_ == i,0], X[kmeans.labels_ == i,1], color=color)
    plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker='^', color='r')
    
plt.title('Clustered Data')
plt.show()
```

```python
np.random.seed(1234)
XX = np.random.randint(-5,5,(5,2))
X_test=XX-np.array([0,2])
print(XX)
print(X_test)
```

```python
y_pred = kmeans.predict(X_test)
print(X_test)
print(y_pred)
```

```python
colors=['blue','black','green']
for color, i in zip(colors, range(3)):
    plt.scatter(X[kmeans.labels_ == i,0], X[kmeans.labels_ == i,1], color=color)
    plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],marker='^')
    plt.scatter(X_test[y_pred==i,0],X_test[y_pred==i,1],color=color,marker='*')
    
plt.title("Clustered data with test data")
plt.show()
```
