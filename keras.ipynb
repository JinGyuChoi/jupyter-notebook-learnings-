```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```python
X=np.array([[0,0],[0,1],[1,0],[1,1]])
Y=np.array([[0],[1],[1],[1]])
```

```python
model=Sequential([
    Dense(input_dim=2, units=1), #인풋2개 아웃풋1개 직접연결(Dense) w1*x1+w2*x2+b
    Activation('sigmoid')
])
```

```python
model.summary() #parameter # : w1,w2,b -> 3개
```

```python
model.compile(loss='binary_crossentropy',optimizer=SGD(lr=0.1)) #SGD:gradiant decsent.
```

```python
classes=model.predict_classes(X,verbose=1)
print('classified: \n',Y==classes)
prob=model.predict_proba(X, verbose=1)
print('output probability: \n', prob)
```

```python
loss=model.evaluate(X,Y,verbose=1)
print("loss:\n",loss)
```

```python
model.fit(X,Y,epochs=200,verbose=1) #training, verbose: 결과 보여주는 여부
```

```python
classes=model.predict_classes(X,verbose=1)
print('classified: \n',Y==classes)
prob=model.predict_proba(X, verbose=1)
print('output probability: \n', prob)
loss=model.evaluate(X,Y,verbose=1)
print("loss:\n",loss)
```

```python
model=Sequential()
model.add(Dense(input_dim=2, units=1))
model.add(Activation('sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer=SGD(lr=0.1))
```

```python
hist=model.fit(X,Y,epochs=200,verbose=2)
```

```python
classes=model.predict_classes(X,verbose=1)
print('classified: \n',Y==classes)
prob=model.predict_proba(X, verbose=1)
print('output probability: \n', prob)
loss=model.evaluate(X,Y,verbose=1)
print("loss:\n",loss)
```

```python
print(hist.history.keys())
```

```python
hist_loss=np.array(hist.history['loss'])
print(hist_loss.shape)
```

```python
plt.plot(hist_loss)
plt.xlabel('the number of epochs')
plt.ylabel('loss')
plt.show()
```
