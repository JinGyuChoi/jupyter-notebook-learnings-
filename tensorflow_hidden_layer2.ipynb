```python
import numpy as np
import tensorflow as tf
```

```python
X=np.array([[0,0],[0,1],[1,0],[1,1]])
Y=np.array([[0],[1],[1],[0]])
```

```python
tf.set_random_seed(123)
x=tf.placeholder(dtype=tf.float32,shape=[None,2])
t=tf.placeholder(tf.float32,shape=[None,1])

#입력층-은닉층
W=tf.Variable(tf.truncated_normal([2,2],seed=0))
b=tf.Variable(tf.zeros([2]))
h=tf.nn.sigmoid(tf.matmul(x,W)+b)

#은닉층-출력층
V=tf.Variable(tf.truncated_normal([2,1],seed=12))
c=tf.Variable(tf.zeros([1]))
y=tf.nn.sigmoid(tf.matmul(h,V)+c)

cross_entropy=-tf.reduce_mean(t*tf.log(y)+(1-t)*tf.log(1-y))

lr=0.1
optimizer=tf.train.GradientDescentOptimizer(lr)
train_step=optimizer.minimize(cross_entropy)

correct_prediction=tf.equal(tf.to_float(tf.greater(y,0.5)),t)
```

```python
init=tf.global_variables_initializer()
sess=tf.Session()
sess.run(init)
print('W = ',sess.run(W))
print('V = ',sess.run(V))
```

```python
for epoch in range(8000):
    sess.run(train_step,feed_dict={
        x:X,
        t:Y
    })
    if epoch%1000==0:
        print('epoch = ',epoch)
```

```python
classified=correct_prediction.eval(session=sess,feed_dict={x:X,t:Y})
prob=y.eval(session=sess,feed_dict={x:X})
print('classified=',classified)
print()
print('output = ',prob)
```

```python
sess.close()
```

```python
#modelling
hidden_dim=10

tf.set_random_seed(1234)

x=tf.placeholder(tf.float32, shape=[None,in_dim])
t=tf.placeholder(tf.float32, shape=[None,out_dim])

#input-hidden
W=tf.Variable(tf.truncated_normal([in_dim,hidden_dim],seed=0))
b=tf.Variable(tf.zeros([hidden_dim]))
h=tf.nn.sigmoid(tf.matmul(x,W)+b)

#hidden-output
V=tf.Variable(tf.truncated_normal([hidden_dim,out_dim],seed=3))
c=tf.Variable(tf.zeros([out_dim]))
y=tf.nn.softmax(tf.matmul(h,V)+c)

cross_entropy=-tf.reduce_mean(tf.reduce_sum(t*tf.log(y)+(1-t),reduction_indices=[1]))

lr=0.05
optimizer=tf.train.GradientDescentOptimizer(lr)
train_step=optimizer.minimize(cross_entropy)

correct_prediction=tf.equal(tf.argmax(y,1),tf.argmax(t,1))
accuracy=tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
```

```python
init=tf.global_variables_initializer()
sess=tf.Session()
sess.run(init)
```

```python
loss_train=cross_entropy.eval(session=sess,feed_dict={x:X_train,t:Y_train})
accuracy_train=accuracy.eval(session=sess, feed_dict={x:X_train,t:Y_train})
print("train_set loss = ",loss_train)
print("\ntrain_set accuracy = ",accuracy_train)
```

```python
n_epoch=2000
for epoch in range(n_epoch):
    sess.run(train_step, feed_dict={
        x:X_train,
        t:Y_train
    })
    if epoch%100==0:
        loss_train=cross_entropy.eval(session=sess,feed_dict={x:X_train,t:Y_train}) #윗줄과 동일
        accuracy_train=accuracy.eval(session=sess,feed_dict={x:X_train,t:Y_train})
        print('epoch = ',epoch,
             '\nloss_train = ',loss_train,
             '\ntrain_set accuracy = ',accuracy_train)
```

```python
loss_train=cross_entropy.eval(session=sess,feed_dict={x:X_train,t:Y_train})
accuracy_train=accuracy.eval(session=sess,feed_dict={x:X_train,t:Y_train})
print("train_set loss = ",loss_train,
     "\ntrain_set accuracy = ",accuracy_train)

y_val_train=y.eval(session=sess,feed_dict={x:X_train, t:Y_train})
print(y_val_train)
print(y_val_train.shape)

classified=correct_prediction.eval(session=sess,feed_dict={x:X_train,t:Y_train})
print(classified)
```

```python
loss_test=cross_entropy.eval(session=sess,feed_dict={x:X_test,t:Y_test})
accuracy_test=accuracy.eval(session=sess,feed_dict={x:X_test,t:Y_test})

print("test_set loss = ",loss_test,
     "\ntest_set accuracy = ",accuracy_test)

y_val_test=y.eval(session=sess,feed_dict={x:X_test,t:Y_test})
print(y_val_test)

classified=correct_prediction.eval(session=sess, feed_dict={x:X_test,t:Y_test})
print(classified)
```
